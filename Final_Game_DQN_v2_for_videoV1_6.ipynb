```python
# CNN-based DQN Model
class DQN(nn.Module):
    def __init__(self, num_actions):
        super(DQN, self).__init__()
        self.conv1 = nn.Conv2d(4, 32, kernel_size=8, stride=4)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=4, stride=2)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, stride=1)
        self.fc1 = nn.Linear(64 * 7 * 7, 512)
        self.fc2 = nn.Linear(512, num_actions)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        return self.fc2(x)
# Preprocess frames (grayscale and resize to 84x84)
# é è™•ç†å½±æ ¼ï¼šè½‰ç‚ºç°éšä¸¦ç¸®æ”¾ç‚º 84x84
def preprocess_frame(frame):
    # frame æ˜¯ numpy array (H, W, 3)ï¼Œå…ˆè½‰ç‚º PIL Image
    # Input is a color image (RGB), convert to PIL format for easier processing.
    # è¼¸å…¥æ˜¯å½©è‰²åœ–åƒï¼ˆRGBï¼‰ï¼Œè½‰æˆ PIL Image ä»¥æ–¹ä¾¿è™•ç†ã€‚
    image = Image.fromarray(frame)

    # è½‰ç°éš
    # Convert the image to grayscale to reduce input complexity.
    # å°‡å½±åƒè½‰ç‚ºç°éšï¼Œé™ä½è¼¸å…¥ç¶­åº¦èˆ‡è¨ˆç®—é‡ã€‚
    image = image.convert('L')

    # resize æˆ 84x84
    # Resize the image to a standard 84x84 shape, as per DQN convention.
    # ä¾ç…§ DQN çš„æ…£ä¾‹å°‡å½±åƒçµ±ä¸€ç¸®æ”¾è‡³ 84x84ã€‚
    image = image.resize((84, 84), Image.Resampling.BILINEAR)  # or NEAREST, or LANCZOS

    # è½‰å› numpy ä¸¦æ­£è¦åŒ–
    # Convert back to NumPy and normalize pixel values to [0, 1].
    # è½‰å› NumPy æ ¼å¼ä¸¦å°‡åƒç´ å€¼æ¨™æº–åŒ–åˆ° [0, 1]ã€‚
    frame = np.asarray(image, dtype=np.float32) / 255.0

    return frame


def stack_frames(stacked_frames, state, is_new_episode):
    # é è™•ç†ç›®å‰å½±æ ¼
    frame = preprocess_frame(state)

    if is_new_episode or stacked_frames is None:
        # If it's a new episode or no previous frames, initialize with 4 identical frames
        # è‹¥æ˜¯æ–°çš„ä¸€é›†æˆ–æ˜¯å°šæœªåˆå§‹åŒ–ï¼Œå‰‡ç”¨ç›®å‰å½±æ ¼è¤‡è£½ 4 æ¬¡å½¢æˆåˆå§‹å †ç–Š
        stacked_frames = deque([frame]*4, maxlen=4)
    else:
        # å¦å‰‡æŠŠæ–°å½±æ ¼åŠ å…¥åˆ°å †ç–Šä¸­ï¼Œè‡ªå‹•æ¨æ£„æœ€èˆŠçš„
        stacked_frames.append(frame)

    # Stack the 4 frames along the first dimension: shape becomes (4, 84, 84)
    # æ²¿è‘—ç¬¬ä¸€ç¶­ï¼ˆchannelï¼‰å †ç–Šæˆ 4 é€šé“è¼¸å…¥ï¼šå½¢ç‹€è®Šæˆ (4, 84, 84)
    stacked_state = np.stack(stacked_frames, axis=0)

    return stacked_state, stacked_frames
```

```python
class SpaceShipEnv():
    def __init__(self, frame_skip=4, stack_frames=4):
        pygame.init()
        # pygame.font.init() # å¦‚æœä¸åœ¨ env ä¸­é¡¯ç¤ºæ–‡å­—ï¼Œå¯ä»¥ä¸ç”¨åˆå§‹åŒ–

        self.screen = None 
        self.clock = pygame.time.Clock()
        self.fps = FPS

        self.game = Game() # éŠæˆ²é‚è¼¯å¯¦ä¾‹

        # å‹•ä½œç©ºé–“: 0: ä¸å‹•, 1: å·¦, 2: å³, 3: å°„æ“Š
        self.action_space = [0, 1, 2, 3] 
        self.action_space_n = len(self.action_space)

        # å¹€è™•ç†ç›¸é—œ
        self.frame_skip = frame_skip
        self.stack_frames = stack_frames
        self.frame_buffer = deque(maxlen=self.stack_frames)

        self.frames_since_last_shot = 0
        self.SHOT_COOLDOWN_FRAMES = 60 # å‡è¨­å­å½ˆå†·å»æ™‚é–“ç‚º 60 å¹€
        
        # å®šç¾©ä¸­é–“å€åŸŸçš„ç¯„åœ
        self.CENTER_ZONE_WIDTH_PERCENT = 0.5 # ä¸­é–“ 20% çš„å€åŸŸ
        self.CENTER_ZONE_X_START = WIDTH * (0.5 - self.CENTER_ZONE_WIDTH_PERCENT / 2)
        self.CENTER_ZONE_X_END = WIDTH * (0.5 + self.CENTER_ZONE_WIDTH_PERCENT / 2)
        # ## æ–°å¢: å‹åˆ©æ¢ä»¶åˆ†æ•¸ ##
        self.WIN_SCORE = 10000 

    def step(self, action):
        """
        åŸ·è¡Œä¸€å€‹å‹•ä½œï¼Œä¸¦å›å‚³ (ä¸‹ä¸€å€‹ç‹€æ…‹, çå‹µ, éŠæˆ²æ˜¯å¦çµæŸ, ç•¶å‰åˆ†æ•¸)
        """
        total_reward = 0.0
        done = False

        # --- ç‹€æ…‹è¿½è¹¤ (State Tracking) ---
        # åœ¨ action åŸ·è¡Œå‰ï¼Œè¨˜éŒ„ç•¶å‰çš„ç‹€æ…‹ï¼Œä»¥ä¾¿å¾ŒçºŒè¨ˆç®—çå‹µ
        prev_health = self.game.player.sprite.health
        prev_score = self.game.score
        prev_is_power = self.game.is_power 
        # --- åŸ·è¡Œå‹•ä½œ (Action Execution) ---
        # æ ¹æ“š frame_skip çš„è¨­å®šï¼Œé‡è¤‡åŸ·è¡ŒåŒä¸€å€‹å‹•ä½œ
        for _ in range(self.frame_skip):
            self.game.update(action) 
            
            # ## å„ªåŒ– ##
            # é™¤ééœ€è¦æ¸²æŸ“ç•«é¢ä¾†ç²å–ç‹€æ…‹ï¼Œå¦å‰‡ draw æ˜¯ä¸å¿…è¦çš„ã€‚
            # é€™è£¡å‡è¨­ self.game.update() æœƒæ›´æ–° self.game.state
            # å¦‚æœä¸æ˜¯ï¼Œå‰‡ä¿ç•™ self.game.draw(None)
            
            # æª¢æŸ¥éŠæˆ²æ˜¯å¦åœ¨ä»»ä½•ä¸€å¹€çµæŸ
            if not self.game.running:
                done = True
                break

        # --- çå‹µè¨ˆç®— (Reward Calculation) ---
        # åœ¨æ‰€æœ‰å¹€éƒ½åŸ·è¡Œå®Œç•¢å¾Œï¼Œæ ¹æ“šã€Œç‹€æ…‹çš„è®ŠåŒ–ã€ä¾†è¨ˆç®—ç¸½çå‹µ
        
        # 1. ä¸»è¦çå‹µï¼šåˆ†æ•¸å¢ç›Š (Score Gain)
        current_score = self.game.score
        current_is_power = self.game.is_power # ç²å–æœ€æ–°çš„ is_power ç‹€æ…‹
        current_health = self.game.player.sprite.health # ç²å–æœ€æ–°çš„å¥åº·å€¼
        current_is_hit_rock = self.game.is_hit_rock # ç²å–æœ€æ–°çš„æ˜¯å¦æ“Šä¸­çŸ³é ­ç‹€æ…‹

        score_gain = current_score - prev_score
        if score_gain > 0:
            # å°‡åˆ†æ•¸çš„å¢ç›Šç›´æ¥è½‰æ›ç‚ºçå‹µï¼Œæ¬Šé‡å¯ä»¥èª¿æ•´
            # é€™æ¨£ Agent æœƒè‡ªç„¶å­¸æœƒå„ªå…ˆæ”»æ“Šåˆ†æ•¸è¼ƒé«˜çš„çŸ³é ­ (é€šå¸¸æ›´å¤§ã€æ›´å±éšª)
            total_reward += score_gain * 0.5  # æ¬Šé‡æ˜¯è¶…åƒæ•¸ï¼Œå¯ä»¥å¯¦é©—èª¿æ•´ ğŸ®
            
            # æ ¹æ“šåˆ†æ•¸å¢ç›Šçµ¦äºˆé¡å¤–çš„å¤§çŸ³é ­çå‹µï¼Œèˆ‡ 1_2 ç‰ˆæœ¬å°é½Š
            if score_gain >= 40: # è¶…å¤§å‹çŸ³é ­ (ä¾‹å¦‚åŠå¾‘ ~40)
                total_reward += 10.0 
            elif score_gain >= 15: # å¤§å‹çŸ³é ­ (ä¾‹å¦‚åŠå¾‘ ~18)
                total_reward += 5.0 
            elif score_gain >= 5: # ä¸­å‹çŸ³é ­
                total_reward += 3 
            else: # å°å‹çŸ³é ­
                total_reward += 2 

        # 2. ä¸»è¦æ‡²ç½°ï¼šç”Ÿå‘½å€¼æå¤± (Health Loss)
        health_diff = current_health - prev_health
        if health_diff < 0: # æ‰è¡€æ‡²ç½°
            total_reward -= abs(health_diff) * 10  # å›ºå®šçš„é«˜é¡æ‡²ç½° ğŸ©¸ (ä½¿ç”¨ abs ä»¥ä¾¿è¨ˆç®—æ­£æ•¸)
        elif health_diff > 0: # è£œè¡€çå‹µ
            total_reward += health_diff * 1.5 # è£œè¡€æœ‰çå‹µï¼ğŸ©¹

        # 3. è¡Œç‚ºå¡‘é€ ï¼šå°„æ“Šæˆæœ¬ (Shooting Cost) åŠæ“Šä¸­çå‹µ
        if action == 3: # å‡è¨­ 3 æ˜¯å°„æ“Š
            if self.frames_since_last_shot < self.SHOT_COOLDOWN_FRAMES:
                total_reward -= 2.5 # å†·å»ä¸­å°„æ“Šæ‡²ç½°å¤§å¹…æé«˜ï¼âŒ
            else:
                self.frames_since_last_shot = 0 # é‡ç½®å°„æ“Šå†·å»è¨ˆæ™‚å™¨
                if current_is_hit_rock and score_gain > 0: # å°„æ“Šä¸¦æˆåŠŸæ“Šä¸­çŸ³é ­ä¸¦å¾—åˆ†
                    total_reward += 10 # æˆåŠŸå°„æ“Šæœ¬èº«æœ‰å°çå‹µ
                else: # äº‚å°„æˆ–æ²’æ“Šä¸­
                    total_reward -= 5 # äº‚å°„æ‡²ç½°å†é«˜ä¸€é» ğŸ¯

        # 4. è¡Œç‚ºå¡‘é€ ï¼šç”Ÿå­˜çå‹µ (Survival Reward)
        if not done:
            if action == 0:
                total_reward += 0.5
            else:
                total_reward += 0.55
        # 5. çµ‚å±€çå‹µ/æ‡²ç½° (Terminal Reward/Penalty)
        if done:
            if current_health <= 0: # å› ç‚ºæ­»äº¡å°è‡´éŠæˆ²çµæŸ
                total_reward -= 500.0 # æ­»äº¡çš„æ‡²ç½°ï¼Œæ¯”å–®æ¬¡å—å‚·æ›´é‡ï¼ğŸ’€
            if current_score >= self.WIN_SCORE: # å‹åˆ©æ¢ä»¶é”æˆ
                total_reward += 10000.0 # å‹åˆ©çš„å–œæ‚…ï¼ğŸ‰

        # åƒåˆ°å¯¶ç‰©çå‹µ
        if current_is_power and not prev_is_power: 
            total_reward += 500 # åƒåˆ°å¯¶ç‰©çµ¦äºˆçå‹µ âœ¨ (èª¿æ•´ç‚º 20 ä»¥èˆ‡ 1_2 æ›´æ¥è¿‘)

        # ä¸­é–“å€åŸŸé¡å¤–çå‹µ (èª¿æ•´ç‚ºèˆ‡ 1_2 ç›¸åŒé‚è¼¯)
        ship_center_x = self.game.player.sprite.rect.centerx 
        if ship_center_x >= self.CENTER_ZONE_X_START and \
           ship_center_x <= self.CENTER_ZONE_X_END: # ä¸å†éœ€è¦åˆ¤æ–·æ˜¯å¦æœ‰æœ€ä½³å°„æ“Šç›®æ¨™å’Œå­å½ˆæº–å‚™å¥½
            total_reward += 0.3 # é¼“å‹µä¿æŒåœ¨ä¸­å¿ƒ ğŸ¯

        # æ–°å¢ï¼šåŸºæ–¼ç¸½åˆ†æ•¸çš„é¡å¤–çå‹µ (æ¯æ¬¡ step éƒ½çµ¦äºˆï¼Œä½†æ¬Šé‡å¯ä»¥å¾ˆå°)
        if self.game.score > 0:
            reward_for_total_score = self.game.score * 0.005 # èˆ‰ä¾‹å€¼ï¼Œéœ€è¦å¯¦é©—èª¿æ•´ï¼ğŸ’¡
            total_reward += reward_for_total_score

        # --- ç‹€æ…‹è™•ç† (State Processing) ---
        # é€™éƒ¨åˆ†é‚è¼¯èˆ‡æ‚¨åŸæœ¬çš„è¨­è¨ˆç›¸åŒï¼Œéå¸¸æ£’ï¼
        current_raw_frame = self.game.state
        processed_frame = preprocess_frame(current_raw_frame)
        self.frame_buffer.append(processed_frame)
        
        # ç¢ºä¿ buffer åœ¨éŠæˆ²ä¸€é–‹å§‹å°±èƒ½å¡«æ»¿
        while len(self.frame_buffer) < self.stack_frames:
            self.frame_buffer.append(processed_frame)
            
        stacked_state = np.array(list(self.frame_buffer))

        # info é€šå¸¸æ˜¯ä¸€å€‹å­—å…¸ï¼Œå¯ç”¨ä¾†å‚³éé™¤éŒ¯è³‡è¨Šï¼Œé€™è£¡æˆ‘å€‘å…ˆç°¡å–®å›å‚³åˆ†æ•¸
        info = {'score': self.game.score}

        return stacked_state, total_reward, done, info

    def reset(self):
        # Reset å‡½æ•¸çš„é‚è¼¯å¾ˆå®Œç¾ï¼Œä¸éœ€è¦å¤§æ”¹
        self.game = Game()
        self.frame_buffer.clear()
        
        # å¡«å……åˆå§‹çš„ frame buffer
        for _ in range(self.stack_frames):
            initial_raw_state = self.game.state 
            processed_frame = preprocess_frame(initial_raw_state) 
            self.frame_buffer.append(processed_frame)
        
        stacked_state = np.array(list(self.frame_buffer))
        return stacked_state

    def render(self):
        # Render å‡½æ•¸çš„é‚è¼¯å¾ˆå®Œç¾ï¼Œä¸éœ€è¦ä¿®æ”¹
        if self.screen is None:
            self.screen = pygame.display.set_mode((WIDTH, HEIGHT))
            pygame.display.set_caption("SpaceShip RL Environment")

        # è®“è¦–çª—å¯ä»¥è¢«é—œé–‰
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                self.close()
                sys.exit()

        self.game.draw(self.screen) 
        pygame.display.update()
        self.clock.tick(self.fps)

    def close(self):
        pygame.quit()
```

```python
import torch
import pygame
import numpy as np
import os
import cv2 # ç‚ºäº†å¯è¦–åŒ–ç‹€æ…‹ï¼Œæˆ‘å€‘å¯èƒ½éœ€è¦å®ƒ

# å¦‚æœæ‚¨é‚„æœ‰å®šç¾© stack_frames æˆ– preprocess_frame å‡½æ•¸ï¼Œç¾åœ¨å¯ä»¥å°‡å®ƒå€‘åˆªé™¤æˆ–è¨»é‡‹æ‰ï¼Œ
# å› ç‚º SpaceShipEnv æ‡‰å·²æ¥ç®¡é€™äº›åŠŸèƒ½ã€‚

# ===================== æ¨¡å‹è¼‰å…¥è¨­ç½® =====================
# è¨­å®š PyTorch é‹è¡Œè¨­å‚™ (GPU æˆ– CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# å‡è¨­é€™è£¡çš„ 'model' è®Šæ•¸å°±æ˜¯æ‚¨è¦è¼‰å…¥çš„ DQN æ¨¡å‹å¯¦ä¾‹
# æˆ‘å€‘ä½¿ç”¨ play_with_trained_model ä¸­çš„ policy_net ä½œç‚ºåƒè€ƒï¼Œçµ±ä¸€è®Šæ•¸å
# å¯¦ä¾‹åŒ–ç’°å¢ƒä»¥ç²å– action_dim
# env_temp = SpaceShipEnv(frame_skip=2, stack_frames=4) # è‡¨æ™‚ç’°å¢ƒï¼Œç”¨æ–¼ç²å–å‹•ä½œç©ºé–“
# action_dim = len(env_temp.action_space) 
# env_temp.close() # é—œé–‰è‡¨æ™‚ç’°å¢ƒ

policy_net = DQN(4).to(device)

# è¼‰å…¥è¨“ç·´å¥½çš„æ¨¡å‹æª”æ¡ˆ
model_path = "checkpointV1_6_2292_1888.pth" # è«‹ç¢ºèªé€™å€‹è·¯å¾‘æ˜¯æ­£ç¢ºçš„ï¼
if not os.path.exists(model_path):
    print(f"éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°æ¨¡å‹æª”æ¡ˆï¼š{model_path} âŒ è«‹ç¢ºèªè·¯å¾‘æ˜¯å¦æ­£ç¢ºï¼")
    exit() # å¦‚æœæ¨¡å‹æª”æ¡ˆä¸å­˜åœ¨ï¼Œç›´æ¥é€€å‡º

# â­â­â­ é—œéµä¿®æ”¹åœ¨é€™è£¡ï¼šè¼‰å…¥æ•´å€‹æª¢æŸ¥é»å­—å…¸ï¼Œç„¶å¾Œæå– policy_net çš„ state_dict â­â­â­
print(f"å˜—è©¦å¾ {model_path} è¼‰å…¥æ¨¡å‹...")
checkpoint = torch.load(model_path, map_location=device)

# æª¢æŸ¥ checkpoint å­—å…¸ä¸­æ˜¯å¦æœ‰ 'policy_net' é€™å€‹éµ
if 'policy_net' in checkpoint:
    policy_net.load_state_dict(checkpoint['policy_net'])
    print("âœ… æˆåŠŸå¾ checkpoint ä¸­è¼‰å…¥ 'policy_net' çš„ state_dictã€‚")
else:
    # å¦‚æœ checkpoint æœ¬èº«å°±æ˜¯ state_dict (ä¸å¤ªå¯èƒ½ï¼Œä½†ä»¥é˜²è¬ä¸€)
    try:
        policy_net.load_state_dict(checkpoint)
        print("âœ… æˆåŠŸè¼‰å…¥æ¨¡å‹ state_dictã€‚ (æ¨¡å‹æª”æ¡ˆå¯èƒ½ç›´æ¥å„²å­˜äº† state_dict)")
    except RuntimeError as e:
        print(f"âŒ ç„¡æ³•å¾ {model_path} è¼‰å…¥æ¨¡å‹ã€‚éŒ¯èª¤è¨Šæ¯ï¼š{e}")
        print("è«‹ç¢ºèª 'checkpoint.pth' æª”æ¡ˆå„²å­˜çš„æ ¼å¼æ˜¯å¦æ­£ç¢ºã€‚")
        exit() # è¼‰å…¥å¤±æ•—å‰‡é€€å‡º

policy_net.eval() # è¨­ç½®ç‚ºè©•ä¼°æ¨¡å¼ (é—œé–‰ Dropout ç­‰)
# =========================================================

# Visualization of trained agent
env = SpaceShipEnv(frame_skip=2, stack_frames=4) # é‡æ–°å¯¦ä¾‹åŒ–ç’°å¢ƒï¼Œç”¨æ–¼éŠæˆ²
# ç”±æ–¼ SpaceShipEnv å…§éƒ¨æœƒè™•ç†æ¸²æŸ“ï¼Œé€™è£¡ä¸éœ€è¦ env.render() æ”¾åœ¨ reset ä¹‹å‰
# å¦‚æœ env.render() åœ¨é€™è£¡æœƒå ±éŒ¯ï¼Œè«‹åˆªé™¤é€™è¡Œæˆ–ç¢ºä¿å®ƒä¸æœƒè§¸ç™¼é¡¯ç¤ºåˆå§‹åŒ–
scores = []
All_games = []
for i in range(100):
    state = env.reset() # ç²å–åˆå§‹ç‹€æ…‹ï¼Œå‡è¨­å®ƒå·²æ˜¯ (4, 84, 84) çš„å †ç–Šå¹€
    done = False
    frames = [] # ç”¨æ–¼å„²å­˜æ¯ä¸€å¹€ç•«é¢ï¼Œä»¥ä¾¿å¾ŒçºŒå¯ä»¥ç”Ÿæˆå½±ç‰‡

    print("ğŸš€ æ¨¡å‹éŠæˆ²å¯è¦–åŒ–é–‹å§‹ï¼æŒ‰ ESC æˆ–é—œé–‰è¦–çª—çµæŸã€‚ğŸš€")

    while not done:
        # è™•ç† Pygame äº‹ä»¶ (ä¾‹å¦‚é—œé–‰è¦–çª—)
        for event in pygame.event.get(): 
            if event.type == pygame.QUIT:
                done = True # è¨­ç½®ç‚º True çµæŸç•¶å‰å›åˆ
                print("ä½¿ç”¨è€…é—œé–‰è¦–çª—ï¼Œå¯è¦–åŒ–æå‰çµæŸã€‚")
                break # è·³å‡ºäº‹ä»¶è™•ç†è¿´åœˆ

        if done: # å¦‚æœå› ç‚ºé—œé–‰è¦–çª—è€Œè·³å‡ºï¼Œå‰‡ç«‹å³çµæŸéŠæˆ²å¾ªç’°
            break

        # æ¸²æŸ“éŠæˆ²ç•«é¢ã€‚ç”±æ–¼ SpaceShipEnv å…§éƒ¨è™•ç†äº†é¡¯ç¤ºï¼Œç›´æ¥å‘¼å«å³å¯ã€‚
        env.render()
        
        with torch.no_grad(): # åœ¨æ¨ç†æ™‚ç¦ç”¨æ¢¯åº¦è¨ˆç®—ï¼Œç¯€çœè¨˜æ†¶é«”å’ŒåŠ é€Ÿ
            # å°‡ NumPy ç‹€æ…‹è½‰æ›ç‚º PyTorch Tensorï¼Œä¸¦å¢åŠ ä¸€å€‹ batch ç¶­åº¦
            state_tensor = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
            
            # å¾æ¨¡å‹ç²å– Q å€¼ (é€™è£¡ä½¿ç”¨ policy_net è€Œä¸æ˜¯ modelï¼Œä¿æŒä¸€è‡´æ€§)
            q_values = policy_net(state_tensor)
            print(q_values)
            # é¸æ“‡ Q å€¼æœ€é«˜çš„å‹•ä½œ
            action = torch.argmax(q_values, dim=1).item()

        # åŸ·è¡Œå‹•ä½œä¸¦ç²å–ä¸‹ä¸€å€‹ç‹€æ…‹ã€çå‹µã€æ˜¯å¦çµæŸå’Œ info
        next_state, reward, done_episode, info = env.step(action)
        
        # å¾ info å­—å…¸ä¸­ç²å–åˆ†æ•¸
        current_score = info.get('score', 0) # ä½¿ç”¨ .get() é¿å… KeyErrorï¼Œå¦‚æœ info ä¸­æ²’æœ‰ 'score'

        # æ›´æ–°ç‹€æ…‹å’Œ done æ¨™èªŒ
        state = next_state
        done = done_episode

        # æŠŠç•«é¢æŠ“ä¸‹ä¾†ï¼ˆRGBï¼‰
        # é€™è£¡å‡è¨­ Pygame çš„ä¸»é¡¯ç¤ºè¡¨é¢æ˜¯é€é pygame.display.get_surface() å–å¾—çš„
        surface = pygame.display.get_surface()
        if surface: # ç¢ºä¿è¡¨é¢å­˜åœ¨
            frame = pygame.surfarray.array3d(surface)  # shape: (W, H, 3)
            frame = np.transpose(frame, (1, 0, 2))      # Pygame æ˜¯ (W, H, C) â†’ imageio/OpenCV æ˜¯ (H, W, C)
            frames.append(frame)
        else:
            print("è­¦å‘Šï¼šç„¡æ³•ç²å– Pygame é¡¯ç¤ºè¡¨é¢ï¼Œå¯èƒ½æœªæ­£ç¢ºåˆå§‹åŒ–æˆ–å·²é—œé–‰ã€‚")
            # å¦‚æœæ²’æœ‰è¡¨é¢ï¼Œè·³éå¹€æ•ç²

    # éŠæˆ²çµæŸå¾Œï¼Œé—œé–‰ç’°å¢ƒ
    # env.close()
    # pygame.quit() # ç¢ºä¿æ‰€æœ‰ Pygame æ¨¡çµ„éƒ½è¢«é—œé–‰

    # æ‰“å°æœ€çµ‚çš„çå‹µå’Œåˆ†æ•¸
    # æ³¨æ„ï¼šé€™è£¡çš„ reward æ˜¯æœ€å¾Œä¸€å€‹ step çš„çå‹µï¼Œä¸æ˜¯ç¸½çå‹µ
    # å¦‚æœæ‚¨éœ€è¦ç¸½çå‹µï¼Œéœ€è¦åœ¨å¾ªç’°ä¸­ç´¯è¨ˆ
    scores.append(current_score) # æœ€çµ‚åˆ†æ•¸
    All_games.append(frames)

for index, score in enumerate(scores):
    print(f"ç¬¬{index+1}å±€éŠæˆ²åˆ†æ•¸: {score}")

# å¦‚æœæ‚¨æƒ³å°‡å¹€å„²å­˜ç‚º GIF æˆ–å½±ç‰‡ï¼Œå¯ä»¥åœ¨é€™è£¡æ·»åŠ é¡å¤–çš„ç¨‹å¼ç¢¼ï¼Œä¾‹å¦‚ä½¿ç”¨ imageio
# import imageio
# output_gif_path = "agent_play.gif"
# imageio.mimsave(output_gif_path, frames, fps=FPS)
# print(f"éŠæˆ²éç¨‹å·²å„²å­˜ç‚º {output_gif_path}")
```
